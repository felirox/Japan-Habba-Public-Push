```python
def send_email(company_name, person_name, person_email, tier, cc=None):
    # me == my email address
    # you == recipient's email address
    me = formataddr(('Japan Habba Public Relations', 'japanhabbapr@gmail.com')) #"japanhabbapr@gmail.com"
    you = formataddr((person_name, person_email)) #"pikachuzhome@gmail.com"

    # Create message container - the correct MIME type is multipart/alternative.
    msg = MIMEMultipart('alternative')
    msg['Subject'] = "Sponsorship Perks and Deliverables for Japan Habba 2022"
    msg['From'] = me
    msg['To'] = you
    
    if cc is not None:
        msg['Cc'] = cc

    # Create the body of the message (a plain-text and an HTML version).
    if tier == 'gold':
        text = gold_plain
        html = gold_html
    elif tier == 'diamond':
        text = diamond_plain
        html = diamond_html
    elif tier == 'platinum':
        text = platinum_plain
        html = platinum_html
    elif tier == 'silver':
        text = silver_plain
        html = silver_html
    else:
        assert False

    # Record the MIME types of both parts - text/plain and text/html.
    part1 = MIMEText(text.format(person_name, company_name), 'plain')
    part2 = MIMEText(html.format(person_name, company_name), 'html')

    # Attach parts into message container.
    # According to RFC 2046, the last part of a multipart message, in this case
    # the HTML message, is best and preferred.
    msg.attach(part1)
    msg.attach(part2)

    path = 'C:/Users/Soham/OneDrive/Documents/'
    filename = 'Suggested JD Format.pdf'

    with open(os.path.join(path, filename), 'rb') as f:
        attachment = MIMEBase('application', 'pdf')
        attachment.set_payload(f.read())
        f.close()
        encoders.encode_base64(attachment)
        attachment.add_header("Content-Disposition", "attachment", filename=filename)
        attachment.add_header('Content-ID', '<{}>'.format(filename))
        msg.attach(attachment)
        
    filename2 = 'Suggested JD Format.docx'

    with open(os.path.join(path, filename2), 'rb') as f2:
        attachment2 = MIMEBase('application', 'vnd.openxmlformats-officedocument.wordprocessingml.document')
        attachment2.set_payload(f2.read())
        f2.close()
        encoders.encode_base64(attachment2)
        attachment2.add_header("Content-Disposition", "attachment", filename=filename2)
        attachment2.add_header('Content-ID', '<{}>'.format(filename2))
        msg.attach(attachment2)

    # Send the message via local SMTP server.
    s = smtplib.SMTP('smtp.gmail.com', 587)

    # start TLS for security
    s.starttls()

    # Authentication
    s.login("japanhabbapr@gmail.com", "jhpr2018")

    # sendmail function takes 3 arguments: sender's address, recipient's address
    # and message to send - here it is sent as one string.
    s.sendmail(me, you, msg.as_string())
    s.quit()
```
